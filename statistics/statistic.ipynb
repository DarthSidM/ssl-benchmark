### mean, variance and standard deviation between multiple random seed

```python
byol_linear_probe = [69.58, 70.00, 70.14]  #40,41,42
simclr_linear_probe = [70.79, 71.50, 71.34] #40,41,42

byol_knn = [67.58, 67.47, 68.66]  #40,41,42
simclr_knn = [66.90, 67.29, 67.36] #40,41,42
```

```python
# mean, variance, std-dev calculations
def compute_stats(data):
    mean = np.mean(data)
    variance = np.var(data)
    std_dev = np.std(data)
    return mean, variance, std_dev

# BYOL Linear Probe Stats
byol_lp_stats = compute_stats(byol_linear_probe)
print("BYOL Linear Probe - Mean: {:.2f}, Variance: {:.4f}, Std Dev: {:.4f}".format(*byol_lp_stats))
# SimCLR Linear Probe Stats
simclr_lp_stats = compute_stats(simclr_linear_probe)
print("SimCLR Linear Probe - Mean: {:.2f}, Variance: {:.4f}, Std Dev: {:.4f}".format(*simclr_lp_stats))

# BYOL KNN Stats
byol_knn_stats = compute_stats(byol_knn)
print("BYOL KNN - Mean: {:.2f}, Variance: {:.4f}, Std Dev: {:.4f}".format(*byol_knn_stats))
# SimCLR KNN Stats
simclr_knn_stats = compute_stats(simclr_knn)
print("SimCLR KNN - Mean: {:.2f}, Variance: {:.4f}, Std Dev: {:.4f}".format(*simclr_knn_stats))

```
